```python
# import relevant libraries

import pandas as pd
import requests
import bs4
```

```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
resp = requests.get(url)
```

```python
soup = bs4.BeautifulSoup(resp.content, "html.parser")
```

```python
data = soup.find("table", id="main_table_countries_today").find("tbody").find_all("tr")
```

```python
data
```

```python
countries = []
infections = []
population = []

# first 8 rows (hidden from view) contain data aggregated by contintent -> exclude them from analysis

for row in data[8:]:
    row_contents = row.find_all("td")
    country_name = row_contents[1].text    

    # number of infections
    if row_contents[2].text.replace(',', '').replace(' ', '').isnumeric():
        country_infections = int(row_contents[2].text.replace(',', '').replace(' ', ''))
    else:
        country_infections = 0
    
    # total population
    if row_contents[14].text.replace(',', '').replace(' ', '').isnumeric():
        country_population = int(row_contents[14].text.replace(',', '').replace(' ', ''))
    else:
        country_population = 0
    
    countries.append(country_name)
    infections.append(country_infections)
    population.append(country_population)
```

```python
covid_dataframe = pd.DataFrame(zip(countries, infections, population), columns=['country', 'num_of_infections', 'total_population'])
covid_dataframe = covid_dataframe.sort_values(by='num_of_infections', ascending=False)
```
