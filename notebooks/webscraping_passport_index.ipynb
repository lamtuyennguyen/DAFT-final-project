```python
# import relevant libraries

import pandas as pd
import requests
import bs4
```

```python
# relevant urls

url_2020_Covid = "https://www.passportindex.org/passport-power-rank-2020-covid19.php"
url_2020 = "https://www.passportindex.org/byRank.php" 
url_2019 = "https://www.passportindex.org/passport-power-rank-2019.php"
url_2018 = "https://www.passportindex.org/passport-power-rank-2018.php"
url_2017 = "https://www.passportindex.org/passport-power-rank-2017.php"
url_2016 = "https://www.passportindex.org/passport-power-rank-2016.php"
url_2015 = "https://www.passportindex.org/passport-power-rank-2015.php"
```

```python
resp_2020_Covid = requests.get(url_2020_Covid)
resp_2020 = requests.get(url_2020)
resp_2019 = requests.get(url_2019)
resp_2018 = requests.get(url_2018)
resp_2017 = requests.get(url_2017)
resp_2016 = requests.get(url_2016)
resp_2015 = requests.get(url_2015)
```

```python
soup_2020_Covid = bs4.BeautifulSoup(resp_2020_Covid.content, "html.parser")
soup_2020 = bs4.BeautifulSoup(resp_2020.content, "html.parser")
soup_2019 = bs4.BeautifulSoup(resp_2019.content, "html.parser")
soup_2018 = bs4.BeautifulSoup(resp_2018.content, "html.parser")
soup_2017 = bs4.BeautifulSoup(resp_2017.content, "html.parser")
soup_2016 = bs4.BeautifulSoup(resp_2016.content, "html.parser")
soup_2015 = bs4.BeautifulSoup(resp_2015.content, "html.parser")
```

```python
data = {}
data['2020_Covid'] = soup_2020_Covid.body.find_all("div", class_ = "allregions")
data['2020'] = soup_2020.body.find_all("div", class_ = "allregions")
data['2019'] = soup_2019.body.find_all("div", class_ = "allregions")
data['2018'] = soup_2018.body.find_all("div", class_ = "allregions")
data['2017'] = soup_2017.body.find_all("div", class_ = "allregions")
data['2016'] = soup_2016.body.find_all("div", class_ = "allregions")
data['2015'] = soup_2015.body.find_all("div", class_ = "allregions")
```

```python
region_name_all_years = {}
score_all_years = {}
region_name = []
score = []

list_of_years = ['2020_Covid','2020', '2019', '2018', '2017', '2016', '2015']

for year in list_of_years:
    for region in data[year]:
        region_name.append(region.find("div", class_ = "name_country").find("span").text)
        score.append(region.find("div", class_ = "name_rank").find("span").text)
    region_name_all_years[year] = region_name[:]
    score_all_years[year] = score[:]
    region_name.clear()
    score.clear()    
    
score_all_years
```

```python
# list of strings 
lst_2020_Covid = zip(region_name_all_years['2020_Covid'], score_all_years['2020_Covid'])
lst_2020 = zip(region_name_all_years['2020'], score_all_years['2020'])
lst_2019 = zip(region_name_all_years['2019'], score_all_years['2019'])
lst_2018 = zip(region_name_all_years['2018'], score_all_years['2018'])
lst_2017 = zip(region_name_all_years['2017'], score_all_years['2017'])
lst_2016 = zip(region_name_all_years['2016'], score_all_years['2016'])
lst_2015 = zip(region_name_all_years['2015'], score_all_years['2015'])

# Calling DataFrame constructor on list 
df_2020_Covid = pd.DataFrame(lst_2020_Covid, columns =['region_name', 'score_2020_Covid'])
df_2020 = pd.DataFrame(lst_2020, columns =['region_name', 'score_2020']) 
df_2019 = pd.DataFrame(lst_2019, columns =['region_name', 'score_2019'])
df_2018 = pd.DataFrame(lst_2018, columns =['region_name', 'score_2018'])
df_2017 = pd.DataFrame(lst_2017, columns =['region_name', 'score_2017'])
df_2016 = pd.DataFrame(lst_2016, columns =['region_name', 'score_2016'])
df_2015 = pd.DataFrame(lst_2015, columns =['region_name', 'score_2015'])
```

```python
df = df_2020_Covid.join(df_2020.set_index('region_name'), on='region_name')\
                  .join(df_2019.set_index('region_name'), on='region_name')\
                  .join(df_2018.set_index('region_name'), on='region_name')\
                  .join(df_2017.set_index('region_name'), on='region_name')\
                  .join(df_2016.set_index('region_name'), on='region_name')\
                  .join(df_2015.set_index('region_name'), on='region_name')
df
```

```python
df.to_csv('../data/passport_index_2015_2020.csv', index=False)
```
